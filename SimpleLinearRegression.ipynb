```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('headbrain.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
X = df['Head Size(cm^3)'].values
y = df['Brain Weight(grams)'].values
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.show()
```

```python
X[:5]
```

```python
mean_x = np.mean(X)
mean_y = np.mean(y)
```

```python
numer = 0
denom = 0

n = len(X)

for i in range(n):
    numer += (y[i] - mean_y) * (X[i] - mean_x)
    denom += (X[i] - mean_x) ** 2

slope = numer / denom
intercept = mean_y - slope * mean_x
```

```python
slope, intercept
```

```python
predictions = []
for i in range(n):
    y_pred = slope * X[i] + intercept
    predictions.append(int(y_pred))
```

```python
predictions[:5]
```

```python
y[:5]
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X, y)
plt.plot(X, predictions, 'r')
plt.show()
```

```python
# MSE - Mean Squared Error
```

```python
error = 0
for i in range(n):
    error += (predictions[i] - y[i]) ** 2

mse = error / n
```

```python
print("MSE :",mse)
```

```python
# RMSE - Root Mean Squared Error
```

```python
np.sqrt(mse)
```
