```python
data = {
    "names" : ["John","Max","Sam","Steve","Nick","Adam","Shawn"],
    "dept" : ["IT","IT","HR","Sales","IT","HR","Sales"],
    "salary" : [45000,50000,25000,80000,35000,50000,35000]
}
```

```python
plt.style.use("ggplot")

plt.figure(figsize=(10,6))
# plt.bar(data["names"], data["salary"])
plt.bar(data["names"], data["salary"], color='red')
plt.title("Salary of each employee")
plt.xlabel("Emp Name")
plt.ylabel("Emp Salary")
plt.show()
```

```python
plt.figure(figsize=(10,6))
# plt.bar(data["names"], data["salary"])
plt.barh(data["names"], data["salary"], color='red')
plt.title("Salary of each employee")
plt.ylabel("Emp Name")
plt.xlabel("Emp Salary")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.pie(data["salary"], labels=data["names"], autopct="%1.2f%%", 
        shadow=True, startangle=90, counterclock=False, explode=(0,0,0.5,0,0,0,0))
plt.title("Salary of each employee")
plt.legend()
plt.show()
```
